```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, fnmatch
```

```python
listOfTestFiles = os.listdir('test')
pattern = "*.csv"
lenArray = []
for file in listOfTestFiles:
    if fnmatch.fnmatch(file, pattern):
        sample = pd.read_csv("test/" + file, dtype={'acoustic_data': np.int16}).reset_index()
        lenArray.append(len(sample))
```

```python
#This is important, because it is the size of data on which we should train our models
```

```python
print(np.unique(np.array(lenArray)))
```

```python
plt.figure(figsize=(13, 10))
plt.plot(sample["acoustic_data"])
plt.title("%0.3f part of sample"% (150000/150000))
plt.ylabel('acoustic data')
plt.xlabel('index')
```

```python
plt.figure(figsize=(13, 10))
plt.plot(sample["acoustic_data"][:20000])
plt.title("%0.3f part of sample"% (20000/150000))
plt.ylabel('acoustic data')
plt.xlabel('index')
```

```python
plt.figure(figsize=(13, 10))
plt.plot(sample["acoustic_data"][:2000])
plt.title("%0.3f part of sample"% (2000/150000))
plt.ylabel('acoustic data')
plt.xlabel('index')
```
